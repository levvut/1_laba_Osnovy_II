# Проанализировать наличие выбросов в данных с помощью диаграмм вида «ящик с усами»;

```python
import matplotlib.pyplot as plt
```

```python
num_cols = {'rent_price', 'area', 'build_area', 'rooms', 'baths', 'garages'}
```

```python
plt.boxplot(cc['rent_price'])
```

#Удалить объекты (строки), соответствующие выбросам

```python
cc = cc.drop(index = cc.index[cc['build_area']>141])
cc = cc.drop(index = cc.index[cc['area']>134])
cc = cc.drop(index = cc.index[cc['rent_price']>(5.2*1000000)])
cc = cc.drop(index = cc.index[cc['rooms']>4.5])
cc = cc.drop(index = cc.index[cc['baths']>3.5])
cc = cc.drop(index = cc.index[cc['garages']>2])
```

```python
cc['rent_price'].value_counts()
```

```python
cc['garages'].value_counts()
```

```python
cc['area'].value_counts()
```

```python
for ax, num_col_name in zip(plt.subplots( 2,3, figsize = (20,20))[1].flatten(), num_cols):
  ax.set_title(num_col_name)
  ax.boxplot(cc[num_col_name])
```

#Выполнить построение диаграммы с помощью функции .jointplot() 


```python
ax = sns.jointplot(x = 'area', y = 'build_area', data = cc, kind = 'reg')
ax.set_axis_labels('Площадь участка', 'Площадь застройки')
```

```python
ax = sns.jointplot(x = 'build_area', y = 'rent_price', data = cc, height=15, ratio=2, marginal_ticks=True, kind = 'reg')
ax.set_axis_labels('Площадь застройки', 'Размер аренды')
```
